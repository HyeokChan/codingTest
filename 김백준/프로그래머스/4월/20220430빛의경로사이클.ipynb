빛의 경로 사이클

각 칸마다 S, L, 또는 R가 써져 있는 격자가 있습니다. 당신은 이 격자에서 빛을 쏘고자 합니다. 이 격자의 각 칸에는 다음과 같은 특이한 성질이 있습니다.
빛이 "S"가 써진 칸에 도달한 경우, 직진합니다.
빛이 "L"이 써진 칸에 도달한 경우, 좌회전을 합니다.
빛이 "R"이 써진 칸에 도달한 경우, 우회전을 합니다.
빛이 격자의 끝을 넘어갈 경우, 반대쪽 끝으로 다시 돌아옵니다. 예를 들어, 빛이 1행에서 행이 줄어드는 방향으로 이동할 경우, 같은 열의 반대쪽 끝 행으로 다시 돌아옵니다.
당신은 이 격자 내에서 빛이 이동할 수 있는 경로 사이클이 몇 개 있고, 각 사이클의 길이가 얼마인지 알고 싶습니다. 경로 사이클이란, 빛이 이동하는 순환 경로를 의미합니다.
예를 들어, 다음 그림은 격자 ["SL","LR"]에서 1행 1열에서 2행 1열 방향으로 빛을 쏠 경우, 해당 빛이 이동하는 경로 사이클을 표현한 것입니다.
ex1.png
이 격자에는 길이가 16인 사이클 1개가 있으며, 다른 사이클은 존재하지 않습니다.
격자의 정보를 나타내는 1차원 문자열 배열 grid가 매개변수로 주어집니다. 주어진 격자를 통해 만들어지는 빛의 경로 사이클의 모든 길이들을 배열에 담아 오름차순으로 정렬하여 return 하도록 solution 함수를 완성해주세요.
제한사항
1 ≤ grid의 길이 ≤ 500
1 ≤ grid의 각 문자열의 길이 ≤ 500
grid의 모든 문자열의 길이는 서로 같습니다.
grid의 모든 문자열은 'L', 'R', 'S'로 이루어져 있습니다.
입출력 예
grid	result
["SL","LR"]	[16]
["S"]	[1,1,1,1]
["R","R"]	[4,4]
입출력 예 설명
입출력 예 #1
문제 예시와 같습니다.
길이가 16인 사이클이 하나 존재하므로(다른 사이클은 없습니다.), [16]을 return 해야 합니다.
입출력 예 #2
주어진 격자를 통해 만들 수 있는 사이클들은 다음 그림과 같습니다.
ex2.png
4개의 사이클의 길이가 모두 1이므로, [1,1,1,1]을 return 해야 합니다.
입출력 예 #3
주어진 격자를 통해 만들 수 있는 사이클들은 다음 그림과 같습니다.
ex3.png
2개의 사이클의 길이가 모두 4이므로, [4,4]를 return 해야 합니다.

1. 지나간 경로에 대한 그래프가 있어야 함.(2차배열)
2. 만약에 경로가 하나라도 겹치면(검색하는 경로인데 지나간 경로 그래프에 있다면 ) 사이클에 포함되는것, 세면 안됨


```python
def solution(grid):
    answer = []
    #새 배열
    g = []
    #우선 배열 생성
    x_len = len(grid)
    y_len = len(grid[0])
    c = dict() #경로체크 딕셔너리
    pos = (0,0)#현재위치
    b_pos = (0,0)#이전위치
    s_pos = (0,0)#시작위치
    direction = ["U", "D", "L", "R"] #빛의 절대 진행방향
    for gr in grid:
        g.append(list(gr))
    
    #모든 격자에서 시작
    for i in range(x_len):
        for j in range(y_len):
            #시작
            #시작지점에서 모든 방향 시작
            for d in direction:
                r = 0#경로길이
                is_full = False
                s_pos = (i,j)
                pos = (i,j)
                b_pos = (i,j)
                md = ""#움직인절대방향
                ftip = ""
                if d == "U":
                    if pos[0] == 0:
                        pos = (x_len-1,pos[1])
                        ftip = "u"
                    else:
                        pos = (pos[0]-1, pos[1])
                    md = "U"
                if d == "D":
                    if pos[0] == x_len-1:
                        pos = (0, pos[1])
                        ftip = "d"
                    else:
                        pos = (pos[0]+1, pos[1])
                    md = "D"
                if d == "L":
                    if pos[1] == 0:
                        pos = (pos[0],y_len-1)
                        ftip = "l"
                    else:
                        pos = (pos[0],pos[1]-1)
                    md = "L"
                if d == "R":
                    if pos[1] == y_len-1:
                        pos = (pos[0],0)
                        ftip = "r"
                    else:
                        pos = (pos[0],pos[1]+1)
                    md = "R"        
                r += 1
                route = str(b_pos[0])+","+str(b_pos[1])+","+str(pos[0])+","+str(pos[1])+ftip
                if c.get(route) == None:
                    c[route] = 1
                else:
                    continue
                #첫 출발 끝
                while True: #경로딕셔너리에 존재할 때 까지 돌림 만약 r이 기존 r 보다 적으면
                    node = g[pos[0]][pos[1]]
                    #print("moved",md,"b_pos",b_pos,"pos",pos,"node",node)
                    b_pos = pos #이전 위치 값 변경
                    tmp = ""
                    tip = ""
                    if node == "S":#현재 도착노드가 S
                        if md == "U":#이전에 움직인 방향이 U
                            if pos[0] == 0:
                                pos = (x_len-1,pos[1])
                                tip = "u"
                            else:
                                pos = (pos[0]-1,pos[1])
                        if md == "D":
                            if pos[0] == x_len-1:
                                pos = (0, pos[1])
                                tip = "d"
                            else:
                                pos = (pos[0]+1, pos[1])
                        if md == "L":
                            if pos[1] == 0:
                                pos = (pos[0], y_len-1)
                                tip = "l"
                            else:
                                pos = (pos[0],pos[1]-1)
                        if md == "R":
                            if pos[1] == y_len-1:
                                pos = (pos[0], 0)
                                tip = "r"
                            else:
                                pos = (pos[0],pos[1]+1)
                        tmp = md #방향 그대로 움직임
                    if node == "L":
                        if md == "U":#왼쪽으로 감
                            if pos[1] == 0:
                                pos = (pos[0], y_len-1)
                                tip = "l"
                            else:
                                pos = (pos[0],pos[1]-1)
                            tmp = "L"
                        if md == "D":#오른쪽으로감
                            if pos[1] == y_len-1:
                                pos = (pos[0], 0)
                                tip = "r"
                            else:
                                pos = (pos[0],pos[1]+1)
                            tmp = "R"
                        if md == "L":#아래로감
                            if pos[0] == x_len-1:
                                pos = (0, pos[1])
                                tip = "d"
                            else:
                                pos = (pos[0]+1,pos[1])
                            tmp = "D"
                        if md == "R":#위로감
                            if pos[0] == 0:
                                pos = (x_len-1, pos[1])
                                tip = "u"
                            else:
                                pos = (pos[0]-1,pos[1])
                            tmp = "U"
                    if node == "R":
                        if md == "U":#오른쪽으로감
                            if pos[1] == y_len-1:
                                pos = (pos[0], 0)
                                tip = "r"
                            else:
                                pos = (pos[0],pos[1]+1)
                            tmp = "R"
                        if md == "D":#왼쪽으로감
                            if pos[1] == 0:
                                pos = (pos[0], y_len-1)
                                tip = "l"
                            else:
                                pos = (pos[0],pos[1]-1)
                            tmp = "L"
                        if md == "L":#위로감
                            if pos[0] == 0:
                                pos = (x_len-1, pos[1])
                                tip = "u"
                            else:
                                pos = (pos[0]-1,pos[1])
                            tmp = "U"
                        if md == "R":#아래로감
                            if pos[0] == x_len-1:
                                pos = (0, pos[1])
                                tip = "d"
                            else:
                                pos = (pos[0]+1,pos[1])
                            tmp = "D"
                    md = tmp
                    r += 1
                    route = str(b_pos[0])+","+str(b_pos[1])+","+str(pos[0])+","+str(pos[1])+tip
                    if c.get(route) == None:
                        c[route] = 1
                    else:
                        if s_pos == b_pos:
                            is_full = True
                        break
                if is_full:
                    answer.append(r-1)
        answer.sort()
    return answer

print(solution(["SL","LR"]))
# [16]
```
